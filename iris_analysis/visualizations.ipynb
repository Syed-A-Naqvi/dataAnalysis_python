```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#This worksheet will cover L21-23. You will use two datasets in this worksheet, diamonds and seaice. Load them as follows:

d_diam = sns.load_dataset('diamonds')
d_ice = sns.load_dataset('seaice')
```

```python
#1. With d_diam, use Seaborn to create a scatterplot with carat (x), price(y), cut (hue).

p = sns.scatterplot(x='carat', y='price', hue='cut', data = d_diam)


#2. Title the plot "The price of diamonds", and label the x and y axes "Weight (carats)" and "Price ($)", respectively.

p = p.set(title='The price of diamonds', xlabel='Weight (carats)', ylabel='Price ($)')
```

```python
#3. With d_ice, use Matplotlib to create a line plot using the Extent column, with a dashed black line

fig = plt.figure()
ax1 = fig.add_subplot()
ax1.plot(d_ice['Extent'], color='black', linestyle='dashed')


#4. Add x-ticks at 0, 3000, 6000, 9000, 12000. Give the ticks labels: "1980", "1992", "2000", "2008", "2016"

ax1.set_xticks([0, 3000, 6000, 9000, 12000])
ax1.set_xticklabels(["1980", "1992", "2000", "2008", "2016"])


#5. Give the x and y-axes labels: "Year", "Thickness (m)" respectively. Give the plot the title "Change in Artic sea ice thickness over time"

ax1.set_xlabel('Year')
ax1.set_ylabel('Thickness (m)')
ax1.set_title('Change in Arctic sea ice thickness over time')
plt.show()
```
